# Observing and testing stationnarity

```python
def plotting(title,data,x,y,xlabel,ylabel):
    fig=px.line(data,x=data[x],y=data[y],labels={x:xlabel,y:ylabel})

    fig.update_layout(template='simple_white',font=dict(size=18),title_text=title,width=650,title_x=0.5,height=400)

    fig.show()

plotting('Truck Sales for Time Series (complete)',database,'Month-Year','Number_Trucks_Sold','Date','Sales')
```

We can observe a clear trend and a certain variance into the data. We can also see a seasonnality, 
but for the purpose of this subject, we will not take account of it.

Testing stationnarity with Augmented Dickey Fuller test :

For this test, we will set the level of certainty needed at 95% following uses. 

If null Hypothesis (H0) accepted for a P-Value equivalent to more than 5% and 10%, the time series is not stationnary

If null Hypothesis rejected (H1), the time series is stationnary

```python
from statsmodels.tsa.stattools import adfuller

def adf_test(series):
    test_results=adfuller(series)
    P_value =  test_results[1]
    if P_value < 0.05:
        print('Data set is Stationnary')
    else:
        print('Dataset is not stationnary')
    print('Adf Statistics : ',test_results[0])
    print('P_Value : ',test_results[1])
    print('Critical Values : ')
    
    for thres, adf_stat in test_results[4].items():
        print('\t%s: %.2f' % (thres,adf_stat))
        
adf_test(database['Number_Trucks_Sold'][1:])
```

As we observe, the test shows the non stationnarity of the time series. 

To get a suitable time series for our models, lets apply some transformations :

# Stationnary Test after transformation

Augmented Dickey Fuller Test for stationnarity 

```python
def adf_test(series):
    test_results=adfuller(series)
    P_value =  test_results[1]
    if P_value < 0.05:
        print('Data set is Stationnary')
    else:
        print('Dataset is not stationnary')
    print('Adf Statistics : ',test_results[0])
    print('P_Value : ',test_results[1])
    print('Critical Values : ')
    
    for thres, adf_stat in test_results[4].items():
        print('\t%s: %.2f' % (thres,adf_stat))
        
print('Logarithm')
adf_test(database['Number_Trucks_Sold_stat_log'][1:])


print('Boxcox')
adf_test(database['Number_Trucks_Sold_stat_boxcox'][1:])
```

With log, H0 is accepted at 5%, the time series are not stationnary, we can apply a second differenciation. to get stationnarity.
With Boxcox, H0 is accpeted at 5% but very close to it, H1 is accepted at 10% though. We can continue with this time series for our example.
